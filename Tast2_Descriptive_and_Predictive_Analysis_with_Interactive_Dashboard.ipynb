## Descriptive Analysis

```python
# converting date column to date time format
df_sales['Date'] = pd.to_datetime(df_sales['Date'], format='%d/%m/%Y')
```

```python
df_sales.head(2)
```

```python
# Basic statistics for 'Weekly_Sales'
sales_stats = df_sales['Weekly_Sales'].describe()
print(sales_stats)

# There are total 4 lack 2 thoushand non-null values in the Weekly sales column,The average sale is around 15k per weeks.
# minus 5 thoushand is total return and refund. 6 lakh is the largest weekly sales
```

```python
# Total sales across the dataset
total_sales = df_sales['Weekly_Sales'].sum()
print(f"Total Sales: {total_sales}")

#Total generated revenue is more than 6 Billion.
```

```python
# Sales by store
sales_by_store = df_sales.groupby('Store')['Weekly_Sales'].sum()
print(sales_by_store)

# this will give us total sales done by each strore.
```

```python
# Sales by department
sales_by_dept = df_sales.groupby('Dept')['Weekly_Sales'].sum()
print(sales_by_dept)

# By above code we can easily find out the total revenue generated by departments.
```

##Predictive Models

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df_sales.set_index('Date', inplace=True)
weekly_sales = df_sales['Weekly_Sales'].resample('W').sum()

# Organizes the data by date, resample method allow us to aggreagate data on different times intervals.
# This will helps us further for analyzing trends over Time.
```

```python
# Prepare features and target
X = pd.DataFrame(weekly_sales.index.astype(int).values, columns=['Date'])
y = weekly_sales.values

# It converted the date index to integers and assigns it to the 'Date' clmn,
# Assign sales value to target variable y
```

#### Train-Test Split

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# 20% data fro testing and rest is for training,
# shuffle preventing data from splitting because here time series data
# X_train is dates , y_train is weekly sales
```

#### Model Building

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

#### Prediction

```python
y_pred = model.predict(X_test)

# using model predicts the sales value for the test set X_test.
```

```python
# Display predicted vs actual sales
predictions = pd.DataFrame({'Date': X_test['Date'], 'Actual Sales': y_test, 'Predicted Sales': y_pred})
print(predictions.head())
```

```python
import dash
from dash import dcc, html
import plotly.express as px
```

```python
# Create a Dash application
app = dash.Dash(__name__)
```

```python
# Create the visualizations
fig1 = px.line(weekly_sales, x=weekly_sales.index, y=weekly_sales.values, title='Weekly Sales')
fig2 = px.bar(df_sales.reset_index(), x='Store', y='Weekly_Sales', title='Sales by Store')
fig3 = px.scatter(df_sales.reset_index(), x='Dept', y='Weekly_Sales', color='IsHoliday', title='Sales by Department (Holiday vs Non-Holiday)')
```

```python
# Define the layout of the app
app.layout = html.Div(children=[
    html.H1(children='Sales Dashboard'),
    # Three graphs are
    dcc.Graph(id='line-chart', figure=fig1),
    dcc.Graph(id='bar-chart', figure=fig2),
    dcc.Graph(id='scatter-chart', figure=fig3)
])
```

```python
# Run the Dash app
if __name__ == '__main__':
    app.run_server(debug=True)
```
